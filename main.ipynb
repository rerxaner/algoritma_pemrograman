```python
import json
import random

# Load messages from the JSON file
with open('messages.json', 'r') as file:
    message_obj = json.load(file)

def get_bot_response(user_message):
    user_message = user_message.lower().strip()

    if user_message in message_obj:
        return message_obj[user_message]
    else:
        bot_message_options = [
            'Maaf, saya tidak mengerti pesanmu.',
            'Silakan tanyakan sesuatu yang lain.',
            'Mohon maaf, saya tidak bisa membantu dengan pertanyaan tersebut.',
            'Saya tidak yakin apa yang Anda maksud.'
        ]
        return random.choice(bot_message_options)

def main():
    print("Chatbot is ready! Type 'exit' to end the chat.")

    while True:
        user_message = input("You: ")

        if user_message.lower() == 'exit':
            print("Chatbot: Goodbye!")
            break

        bot_response = get_bot_response(user_message)
        print(f"Chatbot: {bot_response}")

if __name__ == "__main__":
    main()
```

```python
import json
import random
import tkinter as tk
from tkinter import Scrollbar

# Load messages from the JSON file
with open('messages.json', 'r') as file:
    message_obj = json.load(file)

def get_bot_response(user_message):
    user_message = user_message.lower().strip()

    if user_message in message_obj:
        return message_obj[user_message]
    else:
        bot_message_options = [
            'Maaf, saya tidak mengerti pesanmu.',
            'Silakan tanyakan sesuatu yang lain.',
            'Mohon maaf, saya tidak bisa membantu dengan pertanyaan tersebut.',
            'Saya tidak yakin apa yang Anda maksud.'
        ]
        return random.choice(bot_message_options)

def send_message(event=None):
    user_message = entry.get()
    entry.delete(0, tk.END)

    if user_message.lower().strip() in ['bye', 'exit']:
        add_message("You: " + user_message, 'user')
        add_message("Bot: Goodbye!", 'bot')
        entry.config(state=tk.DISABLED)
        send_button.config(state=tk.DISABLED)
    else:
        add_message("You: " + user_message, 'user')
        bot_response = get_bot_response(user_message)
        add_message("Bot: " + bot_response, 'bot')

def add_message(message, sender):
    # Determine the color scheme based on the sender
    if sender == 'user':
        bg_color = "#dcf8c6"
        text_color = "#075e54"
        justify = 'right'
    else:
        bg_color = "#e2e2e2"
        text_color = "#128c7e"
        justify = 'left'

    # Create chat bubble
    chat_bubble = tk.Label(chat_frame, text=message, wraplength=300, justify=justify, font="Arial 10", bg=bg_color, fg=text_color, padx=10, pady=5)
    chat_bubble.pack(anchor='w' if sender == 'bot' else 'e', padx=10, pady=(5, 0))

# Create GUI window
window = tk.Tk()
window.title("Chatbot")
window.geometry("400x600")

# Chat log frame
chat_frame = tk.Frame(window)
chat_frame.pack(fill=tk.BOTH, expand=True)

# Entry field
entry = tk.Entry(window, bd=0, bg="white", font="Arial")
entry.bind("<Return>", send_message)  # Bind the Enter key to send_message function
entry.pack(fill=tk.X, padx=10, pady=(10, 0))

# Send button
send_button = tk.Button(window, text="Send", command=send_message)
send_button.pack(padx=10, pady=(0, 10), fill=tk.X)

# Start the GUI event loop
window.mainloop()
```
